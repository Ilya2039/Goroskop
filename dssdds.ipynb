```python
from requests_html import AsyncHTMLSession
import os
import nest_asyncio
import asyncio

# Патчинг asyncio для Jupyter
nest_asyncio.apply()

async def fetch_google_results():
    session = AsyncHTMLSession()
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    try:
        response = await session.get("https://www.google.com/search?q=Газпром ИНН", headers=headers)
        await response.html.arender(timeout=20)

        # Проверка содержимого HTML
        if response.html.html:
            # Сохранение в файл
            file_path = os.path.join(os.getcwd(), "google_response.html")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(response.html.html)
            print(f"HTML сохранён в файл '{file_path}'")
        else:
            print("Содержимое HTML пустое.")
    except Exception as e:
        print("Произошла ошибка:", e)

# Вызов асинхронной функции
asyncio.run(fetch_google_results())
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
import os

def fetch_google_results():
    # Настройка драйвера
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Запуск без графического интерфейса
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("start-maximized")
    options.add_argument("disable-infobars")
    options.add_argument("--disable-extensions")
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    )

    # Путь к ChromeDriver
    driver_path = "chromedriver"  # Укажите путь к chromedriver
    service = Service(driver_path)

    # Создание экземпляра браузера
    driver = webdriver.Chrome(service=service, options=options)

    try:
        # Переход на страницу Google
        driver.get("https://www.google.com/")

        # Ввод поискового запроса
        search_box = driver.find_element(By.NAME, "q")
        search_box.send_keys("Газпром")
        search_box.send_keys(Keys.RETURN)

        # Задержка для загрузки результатов
        time.sleep(2)

        # Сбор результатов
        results = driver.find_elements(By.TAG_NAME, "h3")
        for result in results:
            print(result.text)

        # Сохранение HTML-страницы
        html_content = driver.page_source
        with open("google_results.html", "w", encoding="utf-8") as f:
            f.write(html_content)
        print("HTML сохранён в файл 'google_results.html'")
    except Exception as e:
        print("Ошибка:", e)
    finally:
        driver.quit()

# Вызов функции
fetch_google_results()
```
